```python
# Import SQL Alchemy
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.orm import Session
import pandas as pd
```

```python
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
class Data(Base):
    __tablename__ = 'data'
    id = Column(Integer, primary_key=True)
    Year = Column(Integer)
    LocationAbbr = Column(String(255))
    LocationDesc = Column(String(255))
    Question = Column(String(255))
    DataValue = Column(Float)
    StratificationCategory1 = Column(String(255))
    Stratification1 = Column(String(255))
    GeoLocation = Column(String(255))
```

```python
engine = create_engine("sqlite:///diabetes.sqlite")
conn = engine.connect()
```

```python
Base.metadata.create_all(conn)
```

```python
session = Session(bind=engine)
```

```python
x = 1
```

```python
for index, row in df.iterrows():
    data = Data(
        id = x,
        Year=row["Year"],
        LocationAbbr = row["LocationAbbr"],
        LocationDesc = row["LocationDesc"],
        Question = row["Question"],
        DataValue = row["DataValue"],
        StratificationCategory1 = row["StratificationCategory1"],
        Stratification1 = row["Stratification1"],
        GeoLocation = row["GeoLocation"])
    session.add(data)
    session.commit()
    x +=1
```

```python
df = pd.read_sql_query("SELECT * FROM data",conn)
```
